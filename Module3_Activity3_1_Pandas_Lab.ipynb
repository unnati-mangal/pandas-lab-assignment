```python
# Import required libraries
import pandas as pd
# Write code to import pandas and numpy
```

## Task 1: Data Creation & Loading (1 Mark)
Dataset:
    'Name': ['Aman','Riya','Rahul','Sneha','Karan'],
    'Department': ['IT','HR','IT','Finance','HR'],
    'Salary': [50000,60000,55000,65000,52000]

```python
# Create:
# 1. A Pandas Series from a list of 5 elements
print("A Pandas Series from a list of 5 elements")
series = pd.Series([10,20,30,40,50])
print(series)
print("\n")

# 2. A DataFrame with columns: Name, Department, Salary
print("A DataFrame with columns: Name, Department, Salary")
dataset =  {
    'Name': ['Aman','Riya','Rahul','Sneha','Karan'],
    'Department': ['IT','HR','IT','Finance','HR'],
    'Salary': [50000,60000,55000,65000,52000]}
df = pd.DataFrame(dataset)
print(df)
print("\n")

# 3. Write a dataset using (any one): CSV / Excel / JSON
print("Write a dataset using (any one): CSV / Excel / JSON")
df = pd.read_csv("/employee.csv")
print(df)
```

## Task 2: Data Exploration (1 Mark)

```python
# Perform:
# Display first 3 rows
print("Display first 3 rows")
print(df.head(3))
print("\n")
# Display dataset information
print(" Display dataset information")
print(df.info)
print("\n")
# Display statistical summary
print("Display statistical summary")
print(df.describe())
```

## Task 3: Indexing, Filtering & Sorting (1 Mark)

```python
# Perform:
# 1. Extract first 3 rows using .loc[]
print("Extract first 3 rows using .loc[]")
print(df.loc[0:2])
print("\n")
# 2. Extract Salary column using .iloc[]
print("Extract Salary column using .iloc[]")
print(df.iloc[:,2])
print("\n")
# 3. Filter employees with Salary > 55000
print(" Filter employees with Salary > 55000")
filtered_df = df[df['salary'] > 55000]
print(filtered_df)
print("\n")
# 4. Sort dataset by Salary (descending)
print("Sort dataset by Salary (descending)")
sorted_df = df.sort_values(by='salary', ascending=False)
print(sorted_df)
```

## Task 4: Group-by & Aggregation (1 Mark)

```python
# Perform:
# 1. Find average salary department-wise
print("Find average salary department-wise")
avg_salary = df.groupby('department')['salary'].mean()
print(avg_salary)
print("\n")
# 2. Find total and count of employees per department
print("Find total and count of employees per department")
dept_stats = df.groupby('department')['salary'].agg(['sum', 'count'])
print(dept_stats)
```

## Task 5: Data Cleaning & Time-Series (1 Mark)

```python
# Given:
# df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']

# Perform:
# 1. Convert Joining_Date to datetime

# Add Joining_Date column
df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']

# Convert to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
print(df)
print("\n")

# 2. Extract Year and Month

print("year and month:\n")
df['Year'] = df['Joining_Date'].dt.year
df['Month'] = df['Joining_Date'].dt.month
print(df)
print("\n")

# 3. Identify duplicate rows
print( "Identify duplicate rows\n")
duplicates = df[df.duplicated()]
print(duplicates)
print("\n")

# 4. Remove duplicate rows
print("Remove duplicate rows\n")
df_remove = df.drop_duplicates()
print(df_remove)
```
